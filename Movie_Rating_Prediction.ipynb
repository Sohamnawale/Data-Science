```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
column_names = ['MovieNo', 'Title', 'Genres']
```

```python
movies = pd.read_csv('/content/drive/My Drive/movies.csv',sep = '::',names = column_names)
```

```python
Ratings = pd.read_csv('ratings.dat',  sep='::', names= ['MovieID','Rating', 'Timestamp' ])
```

```python
Ratings
```

```python
user = pd.read_csv('users.dat',sep = '::', names=['UserId', 'Sex', 'Age', 'Occupation', 'Zip-code'])
```

```python
user
```

```python
user = pd.DataFrame(user)
ratings = pd.DataFrame(Ratings)

user_ratings = user.merge(ratings,left_on='UserId',right_on='MovieID')
user_ratings
```

```python
movies = pd.DataFrame(movies)

data = user_ratings.merge(movies,left_on='MovieID',right_on='MovieNo')
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
ratings = pd.DataFrame(data.groupby('Title')['Rating'].mean())
ratings
```

```python
ratings['number_of_ratings'] = data.groupby('Title')['Rating'].count()
ratings
```

```python
import seaborn as sns
sns.jointplot(x='Rating',y='number_of_ratings',data=ratings)
```

```python
movie_matrix_UII = data.pivot_table(index='UserId', columns='Title', values='Rating', fill_value=0)
```

```python
movie_matrix_UII = movie_matrix_UII.fillna(0)
```

```python
ratings.sort_values('number_of_ratings',ascending = False).head(10)
```

```python
BacktotheFuture_user_rating = movie_matrix_UII['Back to the Future (1985)']
```

```python
similar_to_BacktotheFuture = movie_matrix_UII.corrwith(BacktotheFuture_user_rating)
```

```python
similar_to_BacktotheFuture.head(2)
```

```python
corr_BacktotheFuture=pd.DataFrame(similar_to_BacktotheFuture,columns = ['correlation'])
corr_BacktotheFuture.dropna(inplace=True)
corr_BacktotheFuture.head()
```

```python
corr_BacktotheFuture = corr_BacktotheFuture.join(ratings['number_of_ratings'])
corr_BacktotheFuture.head()
```

```python
corr_BacktotheFuture[corr_BacktotheFuture['number_of_ratings']>20].sort_values(by='correlation',ascending = False).head()
```
